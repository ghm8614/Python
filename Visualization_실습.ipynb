## 실습 5 : 산점도 그리기

```
# laber을 지정하지 않아도 기본값으로 생성됨
import matplotlib.pyplot as plt
import seaborn as sns

tips = sns.load_dataset('tips')

plt.style.use('default')    # 격자무늬 해제
tips.plot(kind='scatter',x='total_bill',y='tip',s = 10,figsize=(5,3))# s:점크기

plt.title('Scatterplot of Total Bill VS Tip')

plt.show()
```

## 실습 6 : 박스 플롯 그리기

```
import matplotlib.pyplot as plt
import seaborn as sns

tips = sns.load_dataset('tips')  

plt.figure(figsize=(4.5,2.5))
plt.style.use('default')
plt.boxplot(x = [tips[tips['sex']=='Female']['tip'],
                 tips[tips['sex']=='Male']['tip']], labels=['Female','Male'])
plt.title('Boxplot of Tips Sex')

plt.show()
```

## 실습 7 : 회귀선이 있는 산포도 그리기

```
import matplotlib.pyplot as plt
import seaborn as sns

tips = sns.load_dataset('tips')  

plt.figure(figsize=(4,2))
sns.regplot(data=tips, x='size',y='tip',truncate=False)#truncate=False:범위조정

plt.title('Scatterplot of Tip and Headcount')
plt.xlabel('HeadCount')

plt.show()
```

## 실습 8 : 회귀선이 있는 산점도 그리기

```
import matplotlib.pyplot as plt
import seaborn as sns

tips = sns.load_dataset('tips')  

plt.style.use('default')
sns.lmplot(data=tips, x='total_bill',y='tip',hue='smoker',truncate=False,size=3)

plt.title('Scatterplot of Total Bill and Tip')

plt.show()
```

## 실습 9 : 히스토그램&커널 밀도 그래프 그리기

```
import matplotlib.pyplot as plt
import seaborn as sns

tips = sns.load_dataset('tips')  

plt.figure(figsize=(5,3))
sns.distplot(tips['total_bill'])

plt.title('Histogram/Kernel Density Estimation of Total Bill')
plt.ylabel('')   # y축에 공백 만들고 싶을 경우

plt.show()
```

## 실습 10 : 빈도 그래프 그리기

```
import matplotlib.pyplot as plt
import seaborn as sns

tips = sns.load_dataset('tips')  

plt.figure(figsize=(5,3))
sns.countplot(x='day',palette='Set2',data=tips)

plt.title('Count of days')
plt.xlabel('Day of the Week')
plt.ylabel('Frequancy')

plt.show()
```

## 실습 11 : 바이올린 그래프 그리기

```
import matplotlib.pyplot as plt
import seaborn as sns

tips = sns.load_dataset('tips')  

plt.figure(figsize=(5,3))
sns.violinplot(x='time',y='total_bill',data=tips)

plt.title('Violin plot of total bill by time of day')

plt.show()
```

## 실습 12 : 화면 분할하여 히스토그램$커널 밀도 그래프 그리기

```
# subplot을 사용할 것인가 FaceGrid를 사용할 것인가
# 무엇을 기준으로 행과 열로 화면을 분할할 것인가
import matplotlib.pyplot as plt
import seaborn as sns

tips = sns.load_dataset('tips')  

g = sns.FacetGrid(data=tips,col='time',size=3)
g = g.map(sns.distplot,'total_bill') 
# distplot함수의 형태를 가져와서 total_bill값을 넣어라
# g.set_axis_labels('xxx','yyy') : 화면 분할에서 xlabel,ylabel 설정하기

plt.show()
```

## 실습 13 : 화면 분할 산점도 그래프 그리기

```
import matplotlib.pyplot as plt
import seaborn as sns

tips = sns.load_dataset('tips')  

g = sns.FacetGrid(data=tips,col='time',row='smoker',size=2.5)
g = g.map(plt.scatter,'total_bill','tip')

plt.show()
```

```
import matplotlib.pyplot as plt
import seaborn as sns

tips = sns.load_dataset('tips')  

g = sns.FacetGrid(data=tips,col='time',row='smoker',hue='sex',size=2.5)
g = g.map(plt.scatter,'total_bill','tip')
plt.legend()

plt.show()
```
